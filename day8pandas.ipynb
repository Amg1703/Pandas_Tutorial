# CONCATINATING DATAFRAMES 

```python
import pandas as pd 
import numpy as np 
india_weather=pd.DataFrame({
    'Cities':['Chennai','Cochin','Trivandrum'],
    'Temperature':[28,20,18],
    'Windspeed':[7,8,9]
})
india_weather
```

```python
us_weather=pd.DataFrame({
    'Cities':['Boston','LA','New Jersey'],
    'Temperature':[10,22,20],
    'Windspeed':[10,12,14]
})
us_weather
    
```

```python
# pd.concat() - IS USED TO CONCATINATE TWO OR MORE DATAFRAME TOGETHER
# BUT IF YOU NOTICE THE INDEX OF THE 2 DATAFRAME ARE USED SO THE INDEXES ARE REPEATED
concatinated_dataframe=pd.concat([india_weather,us_weather])
concatinated_dataframe
```

```python
# ignore_index = True - IS USED TO IGNORE THE INTIAL INDEX COMING FROM THE 2 DATAFRAME AND USE THE DEFAULT INDEX AND PROPER CONTINUED DATAFRAME INDEXES
concatinated_dataframe=pd.concat([india_weather,us_weather],ignore_index=True)
concatinated_dataframe
```

```python
# keys = [ ] - IS USED TO BE ASSIGNED TO EACH OF THE DATAFRAME SO THAT WE CAN PRINT OR CALL THEM SEPERATELTY FROM THE CONCATINATED DATAFRAME
# BUT keys CANNOT BE USED WHEN ignore_index IS PRESENT AS AN ARGUMENT
concatinated_dataframe=pd.concat([india_weather,us_weather],ignore_index=True,keys=['India Weather','US Weather'])
concatinated_dataframe
```

```python
concatinated_dataframe=pd.concat([india_weather,us_weather], keys=['India Weather','US Weather'])
concatinated_dataframe
```

```python
concatinated_dataframe.loc['India Weather']
```

```python
concatinated_dataframe.loc['US Weather']
```

# TO CONCATINATED AS COLUMNS

```python
temperature_dataframe=pd.DataFrame({
    'Cities':['Chennai','Mumbai','Banglore'],
    'Temperature':[24,21,18]
})
temperature_dataframe
```

```python
# axis=1 - IS USED TO CONCATINATE AS COLUMNS 
# 
windspeed_dataframe=pd.DataFrame({
    'Cities':['Chennai','Mumbai','Banglore'],
    'Windspeed':[10,11,12]
})
windspeed_dataframe
```

```python
# axis=1 - IS USED TO CONCATINATE AS COLUMNS 
# AS BY DEFAULT IT IS axis = 0 SO IT WILL CONCATINATED AS ROWS
column_concatinated_dataframe=pd.concat([temperature_dataframe,windspeed_dataframe],axis=1)
column_concatinated_dataframe
```

```python
# IF INCASE THE ORDER OF THE CITIES WHICH IS THE ROWS ARE ALTERED WE CAN ARRANGE IT USING INDEXES WHILE DEFINING THE DATAFRAME ITSELF
# index = [0,1,2..] IN
temperature_dataframe=pd.DataFrame({
    'Cities':['Chennai','Mumbai','Banglore'],
    'Temperature':[24,21,18]
} ,index=[0,1,2])
windspeed_dataframe=pd.DataFrame({
    'Cities':['Banglore','Mumbai'],
    'Windspeed':[11,12]
}, index=[2,1])
windspeed_dataframe

print(temperature_dataframe)
print(windspeed_dataframe)
```

```python
concat_based_on_index_dataframe=pd.concat([temperature_dataframe, windspeed_dataframe], axis=1, keys=['tempature','windspeed'])
concat_based_on_index_dataframe
```

# PANDAS SERIES

```python
# pd.Series(['Column value1','Column value2',...'Column value N'], name='Column_Name') - TO CREATE A PANDAS SERIES
series=pd.Series(['Humid','Rainy','Sunny'], name='Event')
series
```

```python
concat_dataframe_series=pd.concat([temperature_dataframe,series], axis=1, keys=['temperature&cities','event_series'])
concat_dataframe_series
```

```python
concat_dataframe_series=pd.concat([temperature_dataframe,series], axis=1, keys=['temperature&cities','event_series'])
concat_dataframe_series
```

```python
concat_dataframe_series=pd.concat([temperature_dataframe,series], ignore_index=True)
concat_dataframe_series
```

```python
# WE CAN INDEX BY KEY NAMES FOR DATAFRAMES THAT WAS CONCATINATED BY COLUMNS
concat_dataframe_series.loc['event_series']
```
