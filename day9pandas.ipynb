# MERGE - FOR MERGING TWO OR MORE DATAFRAMES

```python
import pandas as pd 
import numpy as np 

df1=pd.DataFrame({
    'Cities':['Chennai','Mumbai','Banglore'],
    'Temperature':[24,21,18]
})
df1
```

```python
df2=pd.DataFrame({
    'Cities':['Mumbai','Banglore','Trivandrum','Kochin'],
    'windspeed':[24,21,18,19]
})
df2
```

```python
# merge(df1,df2, on='Column you want to link based on') - IS USED MERGE TWO OR MORE DATAFRAMES TOGETHER
# THE merge FUNCTION IS BASED ON SET THEORY INVOLVING UNION AND INNER 
# WHERE BY DEFAULT MERGE DOES IS df1 ^ df2 IT DOES INNER JOIN
# SO ONLY COMMON COLUMN THAT IS ONLY THE COMMON CITIES OF BOTH THE DATAFRAME ARE SELECTED WHICH IS THE INNER JOIN
df3=pd.merge(df1,df2, on='cities')
df3
```

```python
# TO PERFORM df1 U df2 WE NEED TO USE THE how = 'outer' PARAMETER WHEREIN ALL THE ROWS OF THE BOTH DATAFRAME IS JOINED
df4=pd.merge(df1,df2, on='Cities', how='outer')
df4
```

```python
# TO SELECT ONLY COLUMN IN AND COMMON WITH THAT OF FIRST DATAFRAME WE USE how='left' AS THE FIRST DATAFRAME IS 
# CONSIDERED THE LEFT AND THE 2ND DATAFRAME IS CONSIDERED THE RIGHT
df5=pd.merge(df1,df2, on='Cities', how='left')
df5
```

```python
# TO SELECT ONLY COLUMN IN AND COMMON WITH THAT OF 2ND DATAFRAME WE USE how='right' AS THE PARAMETER
df6=pd.merge(df1,df2, on='Cities', how='right')
df6
```

```python
# indicator = True - IS USED TO SHOW WHICH SIDE OF SET THE EACH OF THE ROW CORRESPONDS TO 
df7=pd.merge(df1,df2, on='Cities', how='outer', indicator=True)
df7
```

```python
df1 = pd.DataFrame({
    "Cities": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2=pd.DataFrame({
    "Cities": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
# SO WHEN THE COLUNS ARE COMMON THE MERGE FUNCTION CREATE 2 COLUMNS WITH SUFFIX AS X & Y
df3=pd.merge(df1,df2, on='Cities', how='inner')
df3
```

```python
# suffixes=('_left_', '_right_') - TO CHANGE THE SUFFIX IN THE END OF THE COLUMN NAME
df4=pd.merge(df1,df2, on='Cities', how='inner', suffixes=('_left_', '_right_'))
df4
```
