# HANDLING MISSING DATA

```python
import pandas as pd 
df=pd.read_csv("FillingNANdata.csv", parse_dates=['day'])
df
```

```python
df.set_index('day', inplace=True)
df
```

```python
# fillna 
df_zero=df.fillna(0)
df_zero
```

```python
df_dictionary=df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'No event'
})
df_dictionary
```

```python
# fillna(method='ffill') - TO FILL DATA MISSING VALUES WITH THE DATA ABOVE ITS ROW
df_firstfill=df.fillna(method='ffill')
df_firstfill
```

```python
# fillna(method='bfill') - TO FILL MISSING VALUES WITH THE VALUES AFTER THIS ROW 
df_bfill=df.fillna(method='bfill')
df_bfill
```

```python
# axis='columns' - SO IT TAKES THE VALUE BEFORE IN THE COLUMN IF NO VALUE EXISTS IT ASSIGNS THE VALUE NAN
df_ffill = df.fillna(method='ffill',axis='columns')
df_ffill
```

```python
# axis='rows' - 
df_ffill=df.fillna(method='ffill', axis='rows')
df_ffill
```

```python
# limit = n - IS USED TO LIMIT THE NUMBER OF THE PREVIOUS OR THE AFTER VALUE IS FILLED SO IF IT IS 1 THE PREVIOUS VALUE IS USED TO FILL THE EMPTY VALUE
# ONLY ONCE SO IN THIS OUTPUT WE CAN SEE THAT THE 32.0 OR 7 IN WIND WAS ONLY COPIED AND THE ROW AFTER IT ASSIGNED ONLY AS NAN
df_limit=df.fillna(method='ffill', axis='rows', limit=1)
df_limit
```

```python
# interpolate() - IS USED TO ASSIGN VALUE APPROPRIATELY BASED ON THE DATA VALUE BEFORE AND AFTER IT TO FILL THE NAN VALUE 
# SO FOR THE 2ND ROW IT WAS ASSIGNED THE MIDDLE VALUE BETWEEN 32 ON 1ST ROW AND 28 ON 3RD ROW
# AND IN THE DATE 26TH AND 27TH IT ASSIGNED TO MIDDLE VALUES BETWEEN 32 AND 34 WHICH WAS 32.666 AND 33.333
# THIS IS LINEAR INTERPOLATION
df_interpolate=df.interpolate()
df_interpolate
df
```

```python
df.reset_index(inplace=True)
df
df.set_index('day',inplace=True)
df
df.reset_index(inplace=True)
df=df[['day','temperature','windspeed','event']]
df.set_index('day',inplace=True)
df=df[:9]
df
```

```python
# interpolate(method='time') - IS USED TO INTERPOLATE THE MISSING VALUES OR NAN DATA WITH VALUE BASED ON VALUES BEFORE AND AFTER BUT ALSO CONSIDERING
# THE DATE AS A FACTOR FOR DECIDING FOR EG. IF IT IS 32 ON 1ST AND 28TH ON 5TH THEN MISSING DATA ON 4TH SHOULD BE AROUND 29 AND NOT 30
df_interpolate_time=df.interpolate(method='time')
df_interpolate_time
```

```python
# dropna - TO DROP ROWS WITH NAN VALUES 
# SO ONLY VALUES WITHOUT NAN VALUES ARE PRINTED
df_dropna=df.dropna()
df_dropna
```

```python
# dropna( how='all') - IS USED TO DROP THE ROWS ONLY IF ALL THE VALUES IN THE ROW ARE NAN VALUES
# SO WE CAN SEE THE 27TH DATE WHICH HAS ALL VALUES AS NAN WHICH IS DROPPED AS YOU CAN SEE BUT THE ROWS WITH 1 OR 2 NAN VALUES ARE RETAINED
df_drop_only_all=df.dropna(how='all')
df_drop_only_all
```

```python
#dropna(threshold=N) - IS USED TO SET THE MINIMUM NUMBER OF VALID VALUES (NOT NAN) THAT CAN BE IN THE DATAFRAME
one_NAN_row_df=df.dropna(thresh=1)
one_NAN_row_df
```

```python
# NOW THE THRESHOLD REQUIREMENT IS FOR EACH ROW TO HAVE AT LEAST 2 VALID VALUES SO ONLY 1 NAN VALUE IS ALLOWED 
two_NAN_row_df=df.dropna(thresh=2)
two_NAN_row_df
```

```python
# pd.date_range, pd.DateTimeIndex() , pd.reindex() - IS USED TO FILL THE MISSING DATES IN THE COLUMS
# NOW WE CAN SEE ALL THE DATES ARE ADDED 
date_time=pd.date_range("2025-06-20", "2033-06-09")
dates_to_be_added=pd.DatetimeIndex(date_time)
reindex_df=df.reindex(dates_to_be_added)
reindex_df
```
