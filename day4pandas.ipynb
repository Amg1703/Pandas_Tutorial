```python
import pandas as pd
df=pd.read_csv('SampleDataset.csv')
df
```

```python
# THE FIRST ROW OF THE FILE IS ALWAYS THE HEADER IF IN CASE WE ADD ANOTHER ROW ABOVE THIS ticker ROW THEN THE ADDED ROW ABOVE BECOMES THE HEADER
# 1) skiprows SYNTAX - IN CASE YOU WANT SKIP THE FIRSTT FEW ROWS AND ASSIGN THE HEADER AFTER IT 
df=pd.read_csv('SampleDataset.csv',skiprows=1)
df
```

```python
# 2) header SYNTAX - SIMILAR TO THE PREVIOUS ONE ASSIGNS HEADER ROW USING THE INDEX VALUE
df=pd.read_csv('SampleDataset.csv',header=2)
df
```

```python
df=pd.read_csv('SampleDataset.csv',header=0)
df
```

```python
# 3) na_values SYNTAX - TO CLEAN DATA FIELDS n.a OR not available or in revenue where it is -1 we have to replace the value with NAN 
df=pd.read_csv('SampleDataset.csv', na_values=['n.a.','not available'])
df
```

```python
# BUT IN THE REVENUE COLUMN THERE IS STILL -1 WHICH IS TO BE REPLACED BY NAN BUT THE -1 VALUE IN EPS WILL ALSO BE REPLACED 
df=pd.read_csv('SampleDataset.csv', na_values=['n.a.','not available','-1'])
df
```

```python
# 4) SUPPLY DICTIONARY - IS USED TO TACKLE THE PROBLEM ABOVE WE USE SUPPLY DICTIONARY IN na_values
df=pd.read_csv('SampleDataset.csv', na_values={
    'eps':['not available','n.a.'],
    'revenue':[-1,'not available', 'n.a.'],
    'price':[-1,'not available','n.a.'],
    'people':['not available', 'n.a.']
})
df
```

```python
# TO WRITE A .csv FILE WE USE pd.to_csv("file_name_to_be_saved")
df.to_csv('Newcsv.csv')
```

```python
# TO AVOID INDEXES IN THE SAVED .CSV FILE 
df.to_csv('Newcsvnoindex.csv', index=False)
```

```python
# TO AVOID HEADERS IN THE SAVED .CSV FILE
df.to_csv('Newcsvnoheader.csv', header=False)
```

```python
# CONVERTERS - TO CONVERT THE VALUES IN THE COLUMN OF THE EXCEL FILE

def convert_eps(cell):
    if cell=="not available":
        return None
    return cell

def convert_revenue(cell):
    if cell==-1:
        return None
    return cell

def convert_price(cell):
    if cell=='n.a.':
        return None
    return cell
    
def convert_people(cell):
    if cell=="n.a.":
        return None
    return cell

df=pd.read_excel("SampleDataset.xlsx", converters= { # JUST LIKE SUPPLY DICTIONARY WE DEFINED FOR NA_VALUES WHEREAS THIS WILL CONVERT THE VALUES
    'eps':convert_eps,
    'revenue':convert_revenue,
    'price':convert_price,
    'people':convert_people}
)
df
```

```python
df.to_excel("NewExcel.xlsx")
```

```python
# index=False TO AVOID INDEX 
df.to_excel("NewExcelNoindex.xlsx", index=False)
```

```python
# startrow = n & startcol = n TO START THE ROW AND COLUMNS OF THE DATAFRAME FROM THE MENTIONED NUMBER
df.to_excel("NewExcelShiftedrows&columns.xlsx", index=False, startrow=2, startcol=3)
```
