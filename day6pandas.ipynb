USING REGEX - TO REMOVE THE TEXT IN THE DATASET WE CAN USE REGEX LIKE THE ONE BELOW 

```python
# STEP 1
import pandas as pd
import numpy as np 
df=pd.read_excel("ReplaceDataRegex.xlsx")
df
```

```python
# STEP 2
# FIRST LET'S REMOVE THE SPATIAL VALUES 
replace_df=df.replace({
    'temperature':-99999,
    'windspeed':-88888,
    'event':0
},np.nan)
replace_df
```

```python
#STEP 3
# NOW LETS CONVERT THE NAN VALUES USING FFILL
remove_nan=replace_df.fillna(method='ffill',axis='rows')
remove_nan
```

```python
# STEP 3 
# WE CAN ALSO CONVERT THE NAN VALUES USING INTERPOLATION 
interpolate_df=replace_df.interpolate(method='linear')
interpolate_df
```

```python
# NOW TO REMOVE THE TEXT IN THE DATASET 
# REGEX ARE USED TO DETECT PATTERNS IN THE DATA
# BUT NOW SINCE THE EVENT COLUMN ALSO CONSISTED ALPHABETIC CHARACTERS IT'S ALSO REMOVED
regex_df=remove_nan.replace('[A-Za-z]','',regex=True)
regex_df
```

```python
# TO SOLVE THE ABOVE PROBLEM WE CAN DEFINE DICTIONARY OF COLUMNS FOR THE REGEX ACTION TO BE TAKEN PLACE
regex_df_dictionary=remove_nan.replace({
    'temperature':'[A-Za-z]',
    'windspeed':'[A-Za-z]'
},'',regex=True)
regex_df_dictionary
```

```python
# WE ARE CREATING A DATAFRAME WITH A DICTIONARY
dictionary_df=pd.DataFrame({
    'Name':['Arjun','Sai','Shiva','Ashwin','Shravan','Naresh'],
    'Score':['Exceptional','Good','Average','Poor','Good','Exceptional']
})
dictionary_df
```

```python
# WE CAN ASSIGN AND CONVERT THE SCORES TO NUMERICAL VALUES BYU DEFINING A LIST AND USING THE REPLACE METHOD
numerical_converted_df=dictionary_df.replace(['Exceptional','Good','Average','Poor'],[4,3,2,1])
numerical_converted_df
```
