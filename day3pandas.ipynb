```python
# 3 READING DICTIONARIES AS DATAFRAMES using pd.DataFrame(dictionary_name)
# WHEREIN THE DICTIONARIES ARE DEFINED AS {'key_values':['values']}
dictionary={
    'name':['Arjun','Sai Harish','Shiva'],
    'Age':[21,21,21],
    'Area':['Besant Nagar', 'Adyar', 'Shollinganalur']
}
pd.DataFrame(dictionary)
```

```python
# 4 READING TUPLES AS DATAFRAMES USING pd.DataFrame(tuple_name,columns=['field_name1','field_name2',field_name3']
tuples= [
    ('Ashwin',21, 'Velechery'),
    ('Shravan',21,'Adyar'),
    ('Rohit', 21, 'Adyar')
]
pd.DataFrame(tuples,columns=['Name','Age','Area'])
```

```python
# 5 READING LIST AS A DATAFRAME WHEREIN EACH LIST CONSIST OF COLUMN ALONG WITH VALUE
listname= [
    {'Name':'Arjun','Age':21,'Area':'Besant Nagar'},
    {'Name':'Sai Harish','Age':21,'Area':'Adyar'},
    {'Name':'Shiva','Age':21,'Area':'Shollinganalur'}
]
pd.DataFrame(listname)
```
